```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Walmart.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

Missing values removed to ensure data quality.

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

Duplicate records removed to avoid bias in analysis.

```python
df.describe()
```

```python
df.columns
```

```python
df.info()
```

```python
df['unit_price'] = df['unit_price'].str.replace("$","").astype(float)
```

Unit price column converted to numeric format for analysis.

```python
df.info()
```

```python
df['total'] = df['unit_price']*df['quantity']
```

Total sales column created for revenue analysis.

```python
df.info()
```

```python
df.to_csv('cleaned_data.csv', index=False)
```

```python
from google.colab import files
files.download('cleaned_data.csv')
```

```python
df['total'].sum()
df['total'].describe()
```

```python
df['payment_method'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

Credit Card is the most frequently used payment method based on transaction count, followed closely by Ewallet.

```python
df['total'].plot(kind='hist', bins=30, edgecolor='black')
plt.title('Distribution of Total Sales')
plt.xlabel('Total Amount')
plt.ylabel('Frequency')
plt.show()
```

Most transactions are concentrated in lower value ranges.

```python
df.boxplot(column='total')
plt.title('Outliers in Total Sales')
plt.show()
```

Transactions above ~900 are outliers indicating occasional bulk/high-value purchases

```python
df['category'].value_counts().plot(kind='bar')
plt.title('Category Distribution')
plt.show()
```

Fashion Accessories and Home & Lifestyle dominate sales distribution.


```python
df['time'].value_counts().sort_index().plot()
plt.title('Sales Activity by Time')
plt.show()
```

Peak activity around 15:00 suggests highest customer engagement during afternoon hours.
